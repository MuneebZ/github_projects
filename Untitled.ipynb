# Lower tier local authorities ranked by highest number of charging points per 100,000 population (most recent and time series);

```python
from pandas_ods_reader import read_ods

path = "electric-vehicle-charging-device-statistics-october-2022.ods"


# load a sheet based on its name
sheet_name = "1a"
df_charging_point = read_ods(path, sheet_name)
df_charging_point = df_charging_point.iloc[1:]
df_charging_point.columns = df_charging_point.iloc[0]
# drop the first row
df_charging_point = df_charging_point.drop(df_charging_point.index[0])
df_charging_point.columns
replace_col = [col.replace('[Note 3]','').replace('\n(per 100,000 population)','').replace('July','Jul') for col in df_charging_point.columns]
df_charging_point.columns = replace_col

df_charging_point = df_charging_point[df_charging_point['Local Authority / Region Code'].str.contains('E0')]
df_charging_point
```

```python
[col for col in df_charging_point.columns if "\n(Total Charging Devices)" not in col]
```

```python
for col in [ 'Oct-22 ',
 'Jul-22 ',
 'Apr-22 ',
 'Jan-22 ',
 'Oct-21 ',
 'Jul-21 ',
 'Apr-21 ',
 'Jan-21 ',
 'Oct-20 ',
 'Jul-20 ',
 'Apr-20 ',
 'Jan-20 ',
 'Oct-19 ']:
    
    df_charging_point[col] = pd.to_numeric(df_charging_point[col], errors='coerce')
    
df_charging_point
```

```python
df_charging_point_melt = pd.melt(df_charging_point, id_vars=['Local Authority / Region Code', 'Local Authority / Region Name'], 
value_vars=['Oct-22 ',
 'Jul-22 ',
 'Apr-22 ',
 'Jan-22 ',
 'Oct-21 ',
 'Jul-21 ',
 'Apr-21 ',
 'Jan-21 ',
 'Oct-20 ',
 'Jul-20 ',
 'Apr-20 ',
 'Jan-20 ',
 'Oct-19 '],
        var_name='time', value_name='value')
```

```python
df_charging_point_melt['time'] = pd.to_datetime(df_charging_point_melt['time'], format='%b-%y ')
df_charging_point_melt
```

# Lower tier local authorities ranked by highest number of electric vehicles registered (most recent and time series);

```python
from pandas_ods_reader import read_ods

path = "veh0142.ods"


# load a sheet based on its name
sheet_name = "VEH0142"
df_veh = read_ods(path, sheet_name)


df_veh_new = df_veh.iloc[3:]
df_veh_new.columns = df_veh_new.iloc[0]
# drop the row containing the Units
df_veh_new = df_veh_new.drop(df_veh_new.index[0])

# extract rows where Fuel is in ["Battery electric", "Range extended electric"]
df_veh_new = df_veh_new[df_veh_new['Fuel'].isin(["Battery electric", "Range extended electric"])]

df_veh_new = df_veh_new[df_veh_new['ONS Code [note 6]'].str.contains('E0')]
df_veh_new
```

```python
# Convert all non-float values in the '2022 Q3' column to NaN
df_veh_new['2022 Q3'] = pd.to_numeric(df_veh_new['2022 Q3'], errors='coerce')

for col in ['2022 Q3', '2022 Q2',
       '2022 Q1', '2021 Q4', '2021 Q3', '2021 Q2', '2021 Q1', '2020 Q4',
       '2020 Q3', '2020 Q2', '2020 Q1', '2019 Q4', '2019 Q3', '2019 Q2',
       '2019 Q1', '2018 Q4', '2018 Q3', '2018 Q2', '2018 Q1', '2017 Q4',
       '2017 Q3', '2017 Q2', '2017 Q1', '2016 Q4', '2016 Q3', '2016 Q2',
       '2016 Q1', '2015 Q4', '2015 Q3', '2015 Q2', '2015 Q1', '2014 Q4',
       '2014 Q3', '2014 Q2', '2014 Q1', '2013 Q4', '2013 Q3', '2013 Q2',
       '2013 Q1', '2012 Q4', '2012 Q3', '2012 Q2', '2012 Q1', '2011 Q4',
       '2011 Q3', '2011 Q2', '2011 Q1', '2010 Q4', '2010 Q3', '2010 Q2',
       '2010 Q1', '2009 Q4']:
    df_veh_new[col] = pd.to_numeric(df_veh_new[col], errors='coerce')
```

```python
df_veh_new[['ONS Code [note 6]', 'ONS Geography [note 6]', '2022 Q3']].sort_values(by='2022 Q3', ascending=False)
```

```python
df_veh_new
```

```python
df_vehicle_melt = pd.melt(df_veh_new, id_vars=['ONS Code [note 6]', 'ONS Geography [note 6]'], value_vars=['2022 Q3', '2022 Q2',
       '2022 Q1', '2021 Q4', '2021 Q3', '2021 Q2', '2021 Q1', '2020 Q4',
       '2020 Q3', '2020 Q2', '2020 Q1', '2019 Q4', '2019 Q3', '2019 Q2',
       '2019 Q1', '2018 Q4', '2018 Q3', '2018 Q2', '2018 Q1', '2017 Q4',
       '2017 Q3', '2017 Q2', '2017 Q1', '2016 Q4', '2016 Q3', '2016 Q2',
       '2016 Q1', '2015 Q4', '2015 Q3', '2015 Q2', '2015 Q1', '2014 Q4',
       '2014 Q3', '2014 Q2', '2014 Q1', '2013 Q4', '2013 Q3', '2013 Q2',
       '2013 Q1', '2012 Q4', '2012 Q3', '2012 Q2', '2012 Q1', '2011 Q4',
       '2011 Q3', '2011 Q2', '2011 Q1', '2010 Q4', '2010 Q3', '2010 Q2',
       '2010 Q1', '2009 Q4'],
        var_name='time', value_name='value')
```

```python
df_vehicle_melt
```

```python

# extract the year and quarter information
df_vehicle_melt['year'] = df_vehicle_melt['time'].str[:4].astype(int)
df_vehicle_melt['quarter'] = df_vehicle_melt['time'].str[-1].astype(int)

# create a new datetime column
# create a new date column with the first day of each quarter
df_vehicle_melt['date'] = pd.to_datetime(df_vehicle_melt.apply(lambda x: f"{x['year']}-{(x['quarter']-1) * 3 + 1}-01", axis=1))

# drop the original 'time' column and the year and quarter columns
df_vehicle_melt.drop(['time','year','quarter'], axis=1, inplace=True)
```

```python
df_vehicle_melt = df_vehicle_melt[df_vehicle_melt['ONS Code [note 6]'].str.contains('E0')]
df_vehicle_melt
```
